```python
#Data extraction into simple CSV 
import pandas as pd

# Define the path to your .dat file
file_path = 'dat/liquid_volume_nl12_nx8.dat'

# Initialize lists to store headers and data rows
headers = []
data_rows = []

# Read the file and parse the headers and data
with open(file_path, 'r') as f:
    for line in f:
        if line.startswith('#'):
            # Parse the header line
            headers = list(filter(lambda x: x != '', line.replace('#', '').replace('\n', '').split()))
        else:
            # Parse data lines
            data_rows.append(list(map(float, filter(lambda x: x != '', line.split()))))

# Extract valid column names from the headers
col_names = []
for header in headers:
    parts = header.split(':')
    if len(parts) == 2:
        col_names.append(parts[1].strip())
    else:
        print(f"Skipping invalid header: {header}")

# Convert the list of data rows to a pandas DataFrame
data = pd.DataFrame(data_rows, columns=col_names)

# Print the DataFrame to verify
print(data)

# Optionally, save the DataFrame to a CSV file
data.to_csv('convert_nx/nl12_nx8.csv', index=False)

```

```python

import matplotlib.pyplot as plt
from scipy.signal import welch
import pandas as pd

df = pd.read_csv('convert_nx/long.csv')
time = df['time']
pressure_fluctuation = df['pressure_flc']
average_pressure_fluctuation = pressure_fluctuation.mean()

steady_state_threshold = 0.1  # Define your threshold for steady state
steady_state_time = time[pressure_fluctuation < (average_pressure_fluctuation + steady_state_threshold)].iloc[0]
print("Time to Steady State:", steady_state_time)


plt.figure(figsize=(10, 6))
plt.plot(time, pressure_fluctuation)
plt.axhline(y=average_pressure_fluctuation, color='r', linestyle='--', label='Average Pressure Fluctuation')
plt.title('Pressure Fluctuation vs Time')
plt.xlabel('Time [s]')
plt.ylabel('Pressure Fluctuation')
plt.legend()
plt.grid()
plt.show()
```
